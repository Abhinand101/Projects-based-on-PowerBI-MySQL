Filling the missing values of salary with the mean of the column.

```python
Final=round(Dataset["Salary"].sum()/Dataset["Salary"].notnull().sum())
Dataset["Salary"]=Dataset["Salary"].fillna(Final)
Dataset
```

Column "Height" is having wrong data hence replacing it with proper data.

```python
Dataset["Height"]=np.linspace(150,180,458,dtype=int)
Dataset
```

Question 1

```python
Percentage=Dataset["Team"].value_counts(normalize=True)*100
x=Dataset["Team"].unique()
y=Percentage
plt.figure(figsize=(16,9))
plt.barh(x,y,label="Player Distribution",edgecolor="g")
plt.title("PLAYER DISTRIBUTION BY PERCENTAGE")
plt.legend()
plt.xlabel("PERCENTAGE")
plt.ylabel("TEAM")
plt.show()
```

From the above graph we can see that Boston Celtics has the most no of players out of the total employees in the company 
and Utah Jazz and Portland Trail Blazers have the least.

Question 2

```python
x=Dataset["Position"].unique()
y=Dataset["Position"].value_counts()
plt.bar(x,y,label="Player Distribution",edgecolor="y")
plt.title("PLAYER DISTRIBUTION BY POSITION")
plt.xlabel("POSITION")
plt.ylabel("PLAYERS")
plt.legend()
plt.show()
```

From the above graph we can see that players play most in the PG postion followed closely by SF position,C postion has the least no of players.

Question 3

```python
x=Dataset["Age"].unique()
y=Dataset["Age"].value_counts()
plt.bar(x,y,edgecolor="r",label="Age Distribution")
plt.title("Age Group Distribution")
plt.xlabel("Age")
plt.ylabel("No Of Employees")
plt.legend()
plt.show()
```

From the above graph we can see age group under 20 has the least no of employees and age group between 25 and 30 has the most.

Question 5

```python
Corelation_Datas=Dataset[["Age","Salary"]]
Corelation=Corelation_Datas.corr()
sns.heatmap(Corelation,annot=True,cmap="coolwarm")
plt.title("Age and Salary Correlation Matrix")
plt.show()
```

From the above heatmap we can see that the corelation values is positive,hence we can say that as Age increases the Salary also
increases showing a positive correlation.
